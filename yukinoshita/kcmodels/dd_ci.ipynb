```python
class Equip:
    
    def __init__(self, name: str, karyuku: int=0 , raisou: int=0, kaisyu: int=0) -> None:
        self.__name = name
        self.__karyuku = karyuku
        self.__raisou = raisou
        self.__kaisyu = kaisyu
        self.__type = self.__class__.__name__
        
    @property
    def karyuku(self) -> str:
        return self.__name
        
    @property
    def karyuku(self) -> int:
        return self.__karyuku
    
    @karyuku.setter
    def karyuku(self, value: int) -> None:
        self.__karyuku = value
        
    @property
    def raisou(self) -> int:
        return self.__raisou
    
    @raisou.setter
    def raisou(self, value: int) -> None:
        self.__raisou = value
        
    @property
    def kaisyu(self) -> int:
        return self.__kaisyu
    
    @kaisyu.setter
    def kaisyu(self, value: int) -> None:
        if 0 <= value <= 10:
            self.__kaisyu = value
        raise ValueError('Must less or equal than 10!!')
    
    @property
    def name(self) -> str:
        return self.__name
    
    @property
    def type(self) -> str:
        return self.__type

    
class Mihari(Equip):
    
    def __init__(self, name: str='Mihari', karyuku: int=0, raisou: int=0, kaisyu: int=0,Torpedo: bool=True) -> None:
        super().__init__(name, karyuku, raisou, kaisyu)
        self.__torpedo = Torpedo
        
        
    @property
    def torpedo(self) -> bool:
        return self.__torpedo
    
    @torpedo.setter
    def torpedo(self, value: bool) -> None:
        self.__torpedo = value
        
        
class Radar(Equip):
    def __init__(self, name: str='Radar', karyuku: int=0, raisou: int=0) -> None:
        super().__init__(name, karyuku, raisou)
        
    @property
    def type(self) -> str:
        return self.__type
        
```

```python
class Artillery(Equip):
    
    def __init__(self, name: str, karyuku: int = 0, raisou: int = 0, kaisyu: int = 0) -> None:
        super().__init__(name, karyuku, raisou, kaisyu)
        self.__type = 'Artillery'
        self.__artillery_type = 'Other'
        
    @property
    def type(self) -> str:
        return self.__type
    
    @property
    def art_type(self) -> str:
        return self.__artillery_type 
    
    
class D2(Artillery):
    
    def __init__(self, name: str, karyuku: int = 0, raisou: int = 0, kaisyu: int = 0) -> None:
        super().__init__(name, karyuku, raisou, kaisyu)
        self.__artillery_type = 'D2'
        
        
class D3(Artillery):
    
    def __init__(self, name: str, karyuku: int = 0, raisou: int = 0, kaisyu: int = 0) -> None:
        super().__init__(name, karyuku, raisou, kaisyu)
        self.__artillery_type = 'D3'
        
```

```python
class Torpedo(Equip):
    
    def __init__(self, name: str, karyuku: int = 0, raisou: int = 0, kaisyu: int = 0) -> None:
        super().__init__(name, karyuku, raisou, kaisyu)
        self.__type = 'Torpedo'
        
    @property
    def type(self):
        return self.__type
```

```python
class CutIn:
    
    def __init__(self) -> None:
        self.__type = None
        self.__coefficient = None
        self.__single_rate = None
        
    @property
    def type(self) -> str:
        return self.__type
    
    @property
    def ci_term(self) -> int:
        return self.__coefficient
    
    @property
    def single(self) -> float:
        return self.__single_rate
    
class TorpedoCI(CutIn):
    def __init__(self) -> None:
        self.__type = self.__class__.__name__
        self.__coefficient = 122
        self.__single_rate = 0
        
class TorpedoMihariCI(CutIn):
    def __init__(self) -> None:
        self.__type = self.__class__.__name__
        self.__coefficient = 125
        self.__single_rate = 1/9
        
class ArtilleryTorpedoCI(CutIn):
    def __init__(self) -> None:
        self.__type = self.__class__.__name__
        self.__coefficient = 115
        self.__single_rate = 0
        
class ArtilleryTorpedoRadarCI(CutIn):
    def __init__(self) -> None:
        self.__type = self.__class__.__name__
        self.__coefficient = 115
        self.__single_rate = 1/3
        
class TorpedoRadarMihariCI(CutIn):
    def __init__(self) -> None:
        self.__type = self.__class__.__name__
        self.__coefficient = 140
        self.__single_rate = 1/2
```

```python
import numpy as np

class DD:
    
    def __init__(self, name: str, karyuku :int=0, raisou :int=0, luck :int=0, lv :int=0) -> None:
        self.__name = name.capitalize()
        self.__karyuku = karyuku
        self.__raisou = raisou
        self.__luck = luck
        self.__lv = lv
        self.__equips = []
        
    def __repr__(self) -> str:
        return f'Destroyer {self.__name}'
        
    @property
    def karyuku(self) -> int:
        return self.__karyuku
    
    @karyuku.setter
    def karyuku(self, value: int) -> None:
        self.__karyuku = value
        
    @property
    def raisou(self) -> int:
        return self.__raisou
    
    @raisou.setter
    def raisou(self, value: int) -> None:
        self.__raisou = value
        
    @property
    def luck(self) -> int:
        return self.__luck
    
    @luck.setter
    def luck(self, value: int) -> None:
        self.__luck = value
        
    @property
    def lv(self) -> int:
        return self.__lv
    
    @lv.setter
    def lv(self, value: int) -> None:
        self.__lv = value
        
    @property
    def yasen(self) -> int:
        return self.__karyuku + self.__raisou

    @property
    def ci_term(self) -> int:
        if self.__luck > 50:
            ci_term = 65 + np.sqrt(self.__luck - 50) + 0.8*np.sqrt(self.__lv)
        else:
            ci_term = 15 + self.__luck + 0.75 * np.sqrt(self.__lv)
        return int(ci_term)
    
    @property
    def equips(self) -> list:
        equip_list = []
        for equip in self.__equips:
            equip_list.append(equip.name)
        return equip_list
    
    @equips.setter
    def equips(self, value: Equip) -> None:
        self.__equips.append(value)
```
